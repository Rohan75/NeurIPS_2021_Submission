```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
all_stats = []
baseline = 'DATA/stats'
categories = os.listdir(baseline)
categories.sort()
all_variants = []
for i in range(len(categories)):
    variants = np.array(os.listdir(os.path.join(baseline, categories[i])))
    variants2 = np.array([j[6:] for j in variants])
    try:
        variants2 = np.array([int(j) for j in variants2])
    except:
        pass
    order = np.argsort(variants2)
    variants = variants[order]
    variants2 = variants2[order]
    all_variants.append(variants2)
    all_stats.append([])
    for j in range(len(variants)):
        trials = os.listdir(os.path.join(baseline, categories[i], variants[j]))
        all_stats[i].append([])
        for k in range(len(trials)):
            stat = np.load(os.path.join(baseline, categories[i], variants[j], trials[k]), allow_pickle=True)
            all_stats[i][j].append(stat)
all_stats = np.array(all_stats).transpose([1, 0, 2, 3]).squeeze()
```

```python
correlation_aurocs = np.array(all_stats.transpose([2, 0, 1])[2].tolist())
correlation_mean_aurocs = np.mean(correlation_aurocs, axis=2)

correlation_auprs = np.array(all_stats.transpose([2, 0, 1])[3].tolist())
correlation_mean_auprs = np.mean(correlation_auprs, axis=2)

correlation_snrs = np.array(all_stats.transpose([2, 0, 1])[4].tolist())
correlation_mean_snrs = np.mean(correlation_snrs, axis=2)
```

```python
satori_aurocs = np.array(all_stats.transpose([2, 0, 1])[2+9].tolist())

satori_auprs = np.array(all_stats.transpose([2, 0, 1])[3+9].tolist())

satori_snrs = np.array(all_stats.transpose([2, 0, 1])[4+9].tolist())
```

```python
label_stats = ['Attention AUROC', 'Attention AUPR', 'Attention SNR']
correlation_stats = np.mean([correlation_mean_aurocs, correlation_mean_auprs, correlation_mean_snrs], axis=2).reshape(3, -1)
satori_stats = np.mean([satori_aurocs, satori_auprs, satori_snrs], axis=2).reshape(3, -1)
correlation_stdvs = np.std([correlation_mean_aurocs, correlation_mean_auprs, correlation_mean_snrs], axis=2).reshape(3, -1)
satori_stdvs = np.std([satori_aurocs, satori_auprs, satori_snrs], axis=2).reshape(3, -1)
# statistic, category, trial, variant

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22']
plt.rcParams['ytick.labelsize'] = 10

try:
    os.mkdir('comparison')
except:
    pass

fig = plt.figure(figsize=(4,4))
ax = fig.subplots(1, 1)
for i in range(len(correlation_stats[0])):
    ax.errorbar(correlation_stats[0][i], satori_stats[0][i], correlation_stdvs[0][i], satori_stdvs[0][i], ls='none', label=categories[i], color=colors[i])
ax.scatter(correlation_stats[0], satori_stats[0], label=categories, color=colors)
ax.set_xlim([-0.1,1.1])
ax.set_ylim([-0.1,1.1])
ax.plot([-10, 30], [-10, 30], linestyle='--', color='black')
ax.set_xlabel('GLIFAC', fontsize=20, labelpad=12)
ax.set_title(label_stats[0], fontsize=20, pad=12)
ax.set_ylabel('Local Attention', fontsize=20, labelpad=12)
fig.tight_layout()
fig.savefig(f'comparison/{label_stats[0]}.pdf', dpi=300, format='pdf')

fig = plt.figure(figsize=(4,4))
ax = fig.subplots(1, 1)
for i in range(len(correlation_stats[1])):
    ax.errorbar(correlation_stats[1][i], satori_stats[1][i], correlation_stdvs[1][i], satori_stdvs[1][i], ls='none', label=categories[i], color=colors[i])
ax.scatter(correlation_stats[1], satori_stats[1], label=categories, color=colors)
ax.set_xlim([-0.1,1.1])
ax.set_ylim([-0.1,1.1])
ax.plot([-10, 30], [-10, 30], linestyle='--', color='black')
ax.set_xlabel('GLIFAC', fontsize=20, labelpad=12)
ax.set_title(label_stats[1], fontsize=20, pad=12)
ax.set_ylabel('Local Attention', fontsize=20, labelpad=12)
fig.tight_layout()
fig.savefig(f'comparison/{label_stats[1]}.pdf', dpi=300, format='pdf')

fig = plt.figure(figsize=(4,4))
ax = fig.subplots(1, 1)
for i in range(len(correlation_stats[2])):
    ax.errorbar(correlation_stats[2][i], satori_stats[2][i], correlation_stdvs[2][i], satori_stdvs[2][i], ls='none', label=categories[i], color=colors[i])
ax.scatter(correlation_stats[2], satori_stats[2], label=categories, color=colors)
ax.set_xlim([-0.8,8.8])
ax.set_ylim([-0.8,8.8])
ax.plot([-10, 30], [-10, 30], linestyle='--', color='black')
ax.set_xlabel('GLIFAC', fontsize=20, labelpad=12)
ax.set_title(label_stats[2], fontsize=20, pad=12)
ax.set_ylabel('Local Attention', fontsize=20, labelpad=12)
fig.tight_layout()
fig.savefig(f'comparison/{label_stats[2]}.pdf', dpi=300, format='pdf')

fig = plt.figure(figsize=(3, 2))
ax = fig.subplots(1, 1)
bars = []
for i in range(len(correlation_stats[2])):
    b = ax.errorbar(correlation_stats[2][i], satori_stats[2][i], correlation_stdvs[2][i], satori_stdvs[2][i], ls='none', label=categories[i], color=colors[i])
    bars.append(b)
pts = ax.scatter(correlation_stats[2], satori_stats[2], color=colors)
ax.set_xlim([-0.8,8.8])
ax.set_ylim([-0.8,8.8])
ax.set_xlabel('GLIFAC', fontsize=20, labelpad=12)
ax.set_ylabel('Local Attention', fontsize=20, labelpad=12)
fig.legend(loc='center')
ax.axis('off')
pts.remove()
for i in range(len(bars)):
    bars[i].remove()
fig.tight_layout()
fig.savefig('comparison/legend.pdf', dpi=300, format='pdf')

;
```
