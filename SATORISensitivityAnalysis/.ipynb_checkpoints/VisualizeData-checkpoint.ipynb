```python
import numpy as np
import os
import matplotlib.pyplot as plt

alpha=0.7
```

```python
all_stats = []

baseline = 'DATA/stats'
categories = os.listdir(baseline)
categories.sort()
categories = categories[1:]
for i in range(len(categories)):
    variants = os.listdir(os.path.join(baseline, categories[i]))
    all_stats.append([])
    for j in range(len(variants)):
        trials = os.listdir(os.path.join(baseline, categories[i], variants[j]))
        for k in range(len(trials)):
            stat = np.load(os.path.join(baseline, categories[i], variants[j], trials[k]), allow_pickle=True)
            all_stats[i].append(stat.tolist())
all_stats = np.array(all_stats)
```

```python
ppv_means = np.mean(all_stats.transpose()[0].tolist(), axis=0).squeeze()
ppv_stdvs = np.std(all_stats.transpose()[0].tolist(), axis=0).squeeze()

tpr_means = np.mean(all_stats.transpose()[1].tolist(), axis=0).squeeze()
tpr_stdvs = np.std(all_stats.transpose()[1].tolist(), axis=0).squeeze()
```

```python
try:
    os.mkdir('local_attention_sensitivity')
except:
    pass

fig = plt.figure(figsize=(5,3))
ax = fig.subplots(1,1)

lines = ax.plot(ppv_means.transpose(), label=categories, alpha=alpha, marker='o')
ax.set_xticks(np.arange(10))
ax.set_xticklabels([i/10 for i in range(10)])
ax.set_ylim([-0.1,1.1])
ax.set_ylabel('Attention PPV', fontsize=15)
ax.set_xlabel('Threshold', fontsize=15, labelpad=10)

fig.tight_layout()
fig.savefig('local_attention_sensitivity/PPV.pdf', dpi=300, format='pdf')



fig = plt.figure(figsize=(5,3))
ax = fig.subplots(1,1)

ax.plot(tpr_means.transpose(), alpha=alpha, marker='o')
ax.set_xticks(np.arange(10))
ax.set_xticklabels([i/10 for i in range(10)])
ax.set_ylim([0.45,1.05])
ax.set_ylabel('Attention TPR', fontsize=15)
ax.set_xlabel('Threshold', fontsize=15, labelpad=10)

fig.tight_layout()
fig.savefig('local_attention_sensitivity/TPR.pdf', dpi=300, format='pdf')


fig = plt.figure(figsize=(3, 2))
ax = fig.subplots(1, 1)

ax.plot(tpr_means.transpose(), label=categories, alpha=alpha)
ax.set_xticks(np.arange(10))
ax.set_xticklabels([i/10 for i in range(10)])
ax.set_ylim([0.45,1.05])
ax.set_ylabel('TPR', fontsize=15)
ax.set_xlabel('Threshold', fontsize=15, labelpad=10)
fig.legend(loc='center')
ax.axis('off')
ax.lines = []

fig.tight_layout()
fig.savefig('local_attention_sensitivity/legend.pdf', dpi=300, format='pdf')

;
```
