```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
all_stats = []
baseline = 'CNN10-ReLU/stats'
categories = os.listdir(baseline)
categories.sort()
all_variants = []
for i in range(len(categories)):
    variants = np.array(os.listdir(os.path.join(baseline, categories[i])))
    variants2 = np.array([j[6:] for j in variants])
    try:
        variants2 = np.array([int(j) for j in variants2])
    except:
        pass
    order = np.argsort(variants2)
    variants = variants[order]
    variants2 = variants2[order]
    all_variants.append(variants2)
    all_stats.append([])
    for j in range(len(variants)):
        trials = os.listdir(os.path.join(baseline, categories[i], variants[j]))
        all_stats[i].append([])
        for k in range(len(trials)):
            stat = np.load(os.path.join(baseline, categories[i], variants[j], trials[k]), allow_pickle=True)
            all_stats[i][j].append(stat)
relu_stats = np.array(all_stats)

all_stats = []
baseline = 'CNN10-Exp/stats'
categories = os.listdir(baseline)
categories.sort()
all_variants = []
for i in range(len(categories)):
    variants = np.array(os.listdir(os.path.join(baseline, categories[i])))
    variants2 = np.array([j[6:] for j in variants])
    try:
        variants2 = np.array([int(j) for j in variants2])
    except:
        pass
    order = np.argsort(variants2)
    variants = variants[order]
    variants2 = variants2[order]
    all_variants.append(variants2)
    all_stats.append([])
    for j in range(len(variants)):
        trials = os.listdir(os.path.join(baseline, categories[i], variants[j]))
        all_stats[i].append([])
        for k in range(len(trials)):
            stat = np.load(os.path.join(baseline, categories[i], variants[j], trials[k]), allow_pickle=True)
            all_stats[i][j].append(stat)
exp_stats = np.array(all_stats)


categories = [i.replace('_', ' ') for i in categories]
categories = [i[0].upper() + i[1:].lower() for i in categories]
all_variants[2] = ['BN', 'LN', 'LN2', 'None', 'BN+LN']
all_variants = np.array(all_variants)
relu_stats[2] = relu_stats[2][[3, 0, 1, 2, 4]]
exp_stats[2] = exp_stats[2][[3, 0, 1, 2, 4]]
all_variants[2] = all_variants[2][[3, 0, 1, 2, 4]]
```

```python
relu_ppvs = np.array(relu_stats.transpose([3, 0, 1, 2])[0].tolist()).transpose()[3].transpose()
relu_max_ppvs = np.max(relu_ppvs, axis=2).transpose([0, 2, 1])
relu_mean_ppvs = np.mean(relu_ppvs, axis=2).transpose([0, 2, 1])

relu_tprs = np.array(relu_stats.transpose([3, 0, 1, 2])[1].tolist()).transpose()[3].transpose()
relu_max_tprs = np.max(relu_tprs, axis=2).transpose([0, 2, 1])
relu_mean_tprs = np.mean(relu_tprs, axis=2).transpose([0, 2, 1])

relu_aurocs = np.array(relu_stats.transpose([3, 0, 1, 2])[2].tolist())
relu_max_aurocs = np.max(relu_aurocs, axis=2).transpose([0, 2, 1])
relu_mean_aurocs = np.mean(relu_aurocs, axis=2).transpose([0, 2, 1])

relu_auprs = np.array(relu_stats.transpose([3, 0, 1, 2])[3].tolist())
relu_max_auprs = np.max(relu_auprs, axis=2).transpose([0, 2, 1])
relu_mean_auprs = np.mean(relu_auprs, axis=2).transpose([0, 2, 1])

relu_snrs = np.array(relu_stats.transpose([3, 0, 1, 2])[4].tolist())
relu_max_snrs = np.max(relu_snrs, axis=2).transpose([0, 2, 1])
relu_mean_snrs = np.mean(relu_snrs, axis=2).transpose([0, 2, 1])

relu_auprs = np.array(relu_stats.transpose([3, 0, 1, 2])[-3].tolist()).transpose([0, 2, 1])
relu_coverage = np.array(relu_stats.transpose([3, 0, 1, 2])[16].tolist()).transpose([0, 2, 1])
relu_qvals = -np.log(np.mean(relu_stats.transpose([3, 0, 1, 2])[17].tolist(), axis=-1).transpose([0, 2, 1])+np.finfo(float).eps)



exp_ppvs = np.array(exp_stats.transpose([3, 0, 1, 2])[0].tolist()).transpose()[3].transpose()
exp_max_ppvs = np.max(exp_ppvs, axis=2).transpose([0, 2, 1])
exp_mean_ppvs = np.mean(exp_ppvs, axis=2).transpose([0, 2, 1])

exp_tprs = np.array(exp_stats.transpose([3, 0, 1, 2])[1].tolist()).transpose()[3].transpose()
exp_max_tprs = np.max(exp_tprs, axis=2).transpose([0, 2, 1])
exp_mean_tprs = np.mean(exp_tprs, axis=2).transpose([0, 2, 1])

exp_aurocs = np.array(exp_stats.transpose([3, 0, 1, 2])[2].tolist())
exp_max_aurocs = np.max(exp_aurocs, axis=2).transpose([0, 2, 1])
exp_mean_aurocs = np.mean(exp_aurocs, axis=2).transpose([0, 2, 1])

exp_auprs = np.array(exp_stats.transpose([3, 0, 1, 2])[3].tolist())
exp_max_auprs = np.max(exp_auprs, axis=2).transpose([0, 2, 1])
exp_mean_auprs = np.mean(exp_auprs, axis=2).transpose([0, 2, 1])

exp_snrs = np.array(exp_stats.transpose([3, 0, 1, 2])[4].tolist())
exp_max_snrs = np.max(exp_snrs, axis=2).transpose([0, 2, 1])
exp_mean_snrs = np.mean(exp_snrs, axis=2).transpose([0, 2, 1])

exp_auprs = np.array(exp_stats.transpose([3, 0, 1, 2])[-3].tolist()).transpose([0, 2, 1])
exp_coverage = np.array(exp_stats.transpose([3, 0, 1, 2])[16].tolist()).transpose([0, 2, 1])
exp_qvals = -np.log(np.mean(exp_stats.transpose([3, 0, 1, 2])[17].tolist(), axis=-1).transpose([0, 2, 1])+np.finfo(float).eps)
```

```python
def set_box_color(bp, color):
    plt.setp(bp['boxes'], color=color)
    plt.setp(bp['whiskers'], color=color)
    plt.setp(bp['caps'], color=color)
    plt.setp(bp['medians'], color=color)
```

```python
stat_labels = ['Classification AUPR', 'Attention AUROC', 'Attention AUPR', 'SNR']

plt.rcParams['ytick.labelsize'] = 15
fig = plt.figure(figsize=(5*len(categories), 5*len(stat_labels)))
fig.subplots_adjust(wspace=0.05, hspace=0.05)
ax = fig.subplots(len(stat_labels), len(categories))

for i in range(len(stat_labels)):
    ax[i][0].set_ylabel(stat_labels[i], fontsize=20, labelpad=10)

for category in range(ax.shape[1]):
    ax[len(stat_labels)-1][category].set_title(categories[category], fontsize=20, pad=12, y=-0.25)
    
    relu_bp = ax[0][category].boxplot(relu_auprs[category], widths=0.2, positions=np.arange(relu_auprs[category].shape[1])*1.0-0.15)
    exp_bp = ax[0][category].boxplot(exp_auprs[category], widths=0.2, positions=np.arange(exp_auprs[category].shape[1])*1.0+0.15)
    ax[0][category].set_xticks(np.arange(5))
    ax[0][category].set_xticklabels([])
    ax[0][category].set_ylim([0.91, 1.01])
    set_box_color(relu_bp, 'blue')
    set_box_color(exp_bp, 'green')
    
    relu_bp = ax[1][category].boxplot(relu_mean_aurocs[category], widths=0.2, positions=np.arange(relu_mean_ppvs[category].shape[1])*1.0-0.15)
    exp_bp = ax[1][category].boxplot(exp_mean_aurocs[category], widths=0.2, positions=np.arange(exp_mean_ppvs[category].shape[1])*1.0+0.15)
    ax[1][category].set_xticks(np.arange(5))
    ax[1][category].set_xticklabels([])
    ax[1][category].set_ylim([0.45, 1.05])
    set_box_color(relu_bp, 'blue')
    set_box_color(exp_bp, 'green')
    
    relu_bp = ax[2][category].boxplot(relu_mean_auprs[category], widths=0.2, positions=np.arange(relu_mean_ppvs[category].shape[1])*1.0-0.15)
    exp_bp = ax[2][category].boxplot(exp_mean_auprs[category], widths=0.2, positions=np.arange(exp_mean_ppvs[category].shape[1])*1.0+0.15)
    ax[2][category].set_xticks(np.arange(5))
    ax[2][category].set_xticklabels([])
    ax[2][category].set_ylim([-0.1, 1.1])
    set_box_color(relu_bp, 'blue')
    set_box_color(exp_bp, 'green')
    
    relu_bp = ax[3][category].boxplot(relu_mean_snrs[category], widths=0.2, positions=np.arange(relu_mean_ppvs[category].shape[1])*1.0-0.15)
    exp_bp = ax[3][category].boxplot(exp_mean_snrs[category], widths=0.2, positions=np.arange(exp_mean_ppvs[category].shape[1])*1.0+0.15)
    ax[3][category].set_xticks(np.arange(5))
    ax[3][category].set_xticklabels([])
    ax[3][category].set_ylim([-1, 16])
    set_box_color(relu_bp, 'blue')
    set_box_color(exp_bp, 'green')
    
    ax[-1][category].set_xticklabels(all_variants[category], fontsize=15)
    if category != 0:
        for i in range(len(stat_labels)):
            ax[i][category].set_yticklabels([])
            
fig.savefig('correlation_design_search.pdf', dpi=300, format='pdf')
```
