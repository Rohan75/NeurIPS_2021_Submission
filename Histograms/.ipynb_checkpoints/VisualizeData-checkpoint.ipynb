```python
import numpy as np
import os
import matplotlib.pyplot as plt

alpha = 0.7
```

```python
all_stats = []
baseline = 'DATA/stats'
categories = os.listdir(baseline)
categories.sort()
all_variants = []
for i in range(len(categories)):
    variants = np.array(os.listdir(os.path.join(baseline, categories[i])))
    variants2 = np.array([j[6:] for j in variants])
    try:
        variants2 = np.array([int(j) for j in variants2])
    except:
        pass
    order = np.argsort(variants2)
    variants = variants[order]
    variants2 = variants2[order]
    all_variants.append(variants2)
    all_stats.append([])
    for j in range(len(variants)):
        trials = os.listdir(os.path.join(baseline, categories[i], variants[j]))
        all_stats[i].append([])
        for k in range(len(trials)):
            stat = np.load(os.path.join(baseline, categories[i], variants[j], trials[k]), allow_pickle=True)
            all_stats[i][j].append(stat)
all_stats = np.array(all_stats).squeeze()
```

```python
fig = plt.figure(figsize=(4, 3))
ax = fig.subplots()

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22']

for i in range(len(all_stats)):
    a = ax.hist(all_stats[i], bins=100)
    a[2].remove()
    b = a[0]/np.sum(a[0])
    ax.plot(a[1][:-1], b, color=colors[i], label=categories[i], alpha=alpha)
    
ax.set_xlim([0, 1])
ax.set_ylim([0, 0.2])
# ax.legend(loc='center', frameon=False)
ax.set_xlabel('Attention value', fontsize=15, labelpad=10)
ax.set_ylabel('Frequency', fontsize=15, labelpad=10)

ax.axvline(0.1, linestyle='--', color='black', alpha=alpha)
ax.set_xticks(np.arange(6)/5)
ax.set_yticks(np.arange(5)/4 * ax.get_ylim()[1])

fig.tight_layout()
fig.savefig('attention_distributions.pdf', dpi=300, format='pdf')

;
```

```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
all_stats = []
baseline = 'DATA2/stats'
categories = os.listdir(baseline)
categories.sort()
all_variants = []
for i in range(len(categories)):
    variants = np.array(os.listdir(os.path.join(baseline, categories[i])))
    variants2 = np.array([j[6:] for j in variants])
    try:
        variants2 = np.array([int(j) for j in variants2])
    except:
        pass
    order = np.argsort(variants2)
    variants = variants[order]
    variants2 = variants2[order]
    all_variants.append(variants2)
    all_stats.append([])
    for j in range(len(variants)):
        trials = os.listdir(os.path.join(baseline, categories[i], variants[j]))
        all_stats[i].append([])
        for k in range(len(trials)):
            stat = np.load(os.path.join(baseline, categories[i], variants[j], trials[k]), allow_pickle=True)[1]
            all_stats[i][j].append(stat)
all_stats = np.array(all_stats).squeeze()
```

```python
fig = plt.figure(figsize=(4, 3))
ax = fig.subplots()

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22']

for i in range(len(all_stats)):
    a = ax.hist(all_stats[i], bins=100)
    a[2].remove()
    b = a[0]/np.sum(a[0])
    ax.plot(a[1][:-1], b, color=colors[i], label=categories[i], alpha=alpha)
    sig = np.percentile(all_stats[i], 95)
    ax.axvline(sig, linestyle='--', color=colors[i], alpha=alpha)
    
ax.set_xlim([0, 1])
ax.set_ylim([0, 0.2])
# ax.legend(loc='center', frameon=False)
ax.set_xlabel('Attention value', fontsize=15, labelpad=10)
ax.set_ylabel('Frequency', fontsize=15, labelpad=10)

ax.set_xticks(np.arange(6)/5)
ax.set_yticks(np.arange(5)/4 * ax.get_ylim()[1])

fig.tight_layout()
fig.savefig('background_distributions.pdf', dpi=300, format='pdf')

;
```
